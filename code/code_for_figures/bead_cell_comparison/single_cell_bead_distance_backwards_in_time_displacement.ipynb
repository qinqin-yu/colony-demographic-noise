```python
import numpy
import matplotlib.pyplot as plt
import math
import pandas as pd
plt.rcParams.update({'font.size': 18})
```

```python
beads_filename = '/Users/qinqinyu/Documents/hallatschek_lab/data_changed/20200120_olympus_single_cell_beads/20200123_bw25113_unwashed_1um_delt2min_delz1um_18h-21h/tracks/beads/01.csv'
df_beads = pd.read_csv(beads_filename, index_col = 0)
plt.plot(df_beads['X'], df_beads['Y'], 'k')

d_all = numpy.empty((len(df_beads), 14))
for i in range(1,15):
    if i<10:
        cells_filename = '/Users/qinqinyu/Documents/hallatschek_lab/data_changed/20200120_olympus_single_cell_beads/20200123_bw25113_unwashed_1um_delt2min_delz1um_18h-21h/tracks/cells/0'+str(i)+'.csv'
    else:
        cells_filename = '/Users/qinqinyu/Documents/hallatschek_lab/data_changed/20200120_olympus_single_cell_beads/20200123_bw25113_unwashed_1um_delt2min_delz1um_18h-21h/tracks/cells/'+str(i)+'.csv'
    df_cells = pd.read_csv(cells_filename, index_col = 0)
    plt.plot(df_cells['X'], df_cells['Y'])
    d = numpy.sqrt((df_beads['X']-df_cells['X'])**2 + (df_beads['Y']-df_cells['Y'])**2)
    d_all[:,i-1] = d
    
d_all = d_all[:31, :]
d_all.shape
```

```python
colors = numpy.array([[0, 0.4470, 0.7410], 
                      [0.8500, 0.3250, 0.0980],
                      [0.9290, 0.6940, 0.1250],
                      [0.4940, 0.1840, 0.5560],
                      [0.4660, 0.6740, 0.1880],
                      [0.3010, 0.7450, 0.9330],
                      [0.6350, 0.0780, 0.1840]])
colors_index = 0
for i in [0, 4, 5, 6, 13]:#range(d_all.shape[1]):
    plt.plot((d_all[:,i]-d_all[-1,i])/1.7, color = colors[colors_index,:])
    #if i<d_all.shape[1]-1 and d_all[0,i+1]!=d_all[0,i]:
    colors_index += 1
# plt.plot([0, d_all.shape[0]], [0, 0], color = 'k')
plt.fill_between([0, 40], -1, 1, facecolor='gray', alpha = 0.2)
plt.plot([0,40], [0, 0], 'k')
plt.xticks(ticks = numpy.array([0,10,20,30,40]), labels = numpy.array(['20h', '20h20m', '20h40m', '21h00']))
plt.xlabel('backwards in time')
plt.ylabel('deviation from final distance \n to bead [effective cell size]')

plt.xlim([0, d_all.shape[0]-1])
plt.ylim([-2,2])

plt.gca().invert_xaxis()

plt.savefig('/Users/qinqinyu/Documents/hallatschek_lab/data_changed/20200228_spatial_genetic_drift_summary/figures/bead_cell_deviation1_backwards_neighbors_cell_size.pdf')
plt.show()
```

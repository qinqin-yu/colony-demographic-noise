```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob, os
import csv
import seaborn as sns
from scipy.special import erf

sns.set(style="whitegrid",font_scale=1.2)
```

```python
def import_area_imagej(filename):
    csvfile = open(filename, 'r')
    csvreader = csv.reader(csvfile, delimiter = ",")
    inoculation_size_list = list(csvreader)
    inoculation_size_all = np.array(inoculation_size_list)
    header = inoculation_size_all[0,:]
    label = inoculation_size_all[1:,1]
    area = inoculation_size_all[1:,2].astype('int')

    im_num = np.empty(len(label)).astype('int')
    for i in range(len(label)):
        ind = label[i].find(':')
        ind2 = label[i].find('_')
        im_num[i] = int(label[i][ind+1:ind2])

    area_rearray = np.empty(96)
    area_rearray.fill('nan')
    area_rearray[im_num] = area
    return area_rearray
```

```python
# Specify paths

fitness_path = '/Users/qinqinyu/Dropbox/keio_data/nonneutral_data/fitness/'
sectors_path = '/Users/qinqinyu/Dropbox/keio_data/nonneutral_data/nonneutral_sector_number/'
initial_fraction_path = '/Users/qinqinyu/Dropbox/keio_data/nonneutral_data/nonneutral_mixture_initial_fraction/'
inoculum_path = '/Users/qinqinyu/Dropbox/keio_data/nonneutral_data/nonneutral_mixture_inoculum_area/'

genotypes_filename = '/Users/qinqinyu/Dropbox/keio_data/nonneutral_data/genotypes_order.csv'
genotypes = pd.read_csv(genotypes_filename, names = ['genotype'])
```

```python
# Resave fitness data

fitness_all = pd.DataFrame(columns = ['genotype', 'cm', 'replicate', 's'])

os.chdir(fitness_path)
fitness_filenames = []
for file in glob.glob('s_*'):
    fitness_filenames.append(file)

letters = 'ABC'

for fitness_filename in fitness_filenames: 
    excluded_filename = 'excluded_' + fitness_filename[fitness_filename.find('s_')+2:fitness_filename.find('.')] + '.csv'
    fitness = pd.read_csv(fitness_path + fitness_filename, names = ['s'])
    excluded = pd.read_csv(fitness_path + excluded_filename)
    fitness.loc[fitness.index.isin(excluded['collision number'])] = np.nan
    
    letter = fitness_filename[fitness_filename.find('s_')+2]
    letter_idx = letters.find(letter)
    
    genotype_array = np.transpose(np.tile(np.transpose(genotypes[letter_idx*4:letter_idx*4+4].values), 8))
    fitness['genotype'] = genotype_array

    cm = fitness_filename[fitness_filename.find('cm')+2]
    fitness['cm'] = cm
    fitness['replicate'] = np.repeat(np.array(list(range(8))),4)
    fitness_all = fitness_all.append(fitness)
#     print(fitness_all)

fitness_all.reset_index(inplace = True, drop = True)
```

```python
fitness_all.sort_values(by = ['genotype', 'cm', 'replicate'], inplace = True, ignore_index = True)
# fitness_all.to_csv('/Users/qinqinyu/Dropbox/keio_data/nonneutral_data/fitness_all.csv')
```

```python
# Resave establishment probability data
conv = 0.1240 #0.1240 pixels = 1 um for 8x;

pest_all = pd.DataFrame(columns = ['genotype', 'cm', 'p0_expected', 'p0_actual', 'p0_actual_err', 'replicate', 'pest', 'pest_err'])

print(pest_all)
# sectors

os.chdir(sectors_path)
sectors_filenames = []
for file in glob.glob('*.csv'):
    sectors_filenames.append(file)
    
for sectors_filename in sectors_filenames:
    # sectors_filename = 'p_0_002_0ug_mL_cm_number_sectors.csv'
    sectors = pd.read_csv(sectors_path + sectors_filename)['number of sectors'].values
    p0 = '0.' + sectors_filename[sectors_filename.find('p_')+4:sectors_filename.find('p_')+7]
    cm = sectors_filename[sectors_filename.find('ug_mL')-1:sectors_filename.find('ug_mL')]

    # areas
    inoculum_filename = 'inoculum_area_p0_' + p0[2:] + '_cm0.csv'
    areas = import_area_imagej(inoculum_path + inoculum_filename)/(conv**2)

    # initial fraction
    initial_fraction_filename = 'frac_yfp_manually_changed_p0_' + p0[2:] + '.csv'
    initial_fraction_df = pd.read_csv(initial_fraction_path + initial_fraction_filename)
    initial_fraction = np.tile(initial_fraction_df['frac_yfp'].values, 8)
    
    cfp_cfus = initial_fraction_df['num_cfp']
    yfp_cfus = initial_fraction_df['num_yfp']
    total_cfus = cfp_cfus + yfp_cfus
    total_cfus = np.tile(total_cfus, 8)
    
    initial_fraction_err = np.sqrt(initial_fraction*(1-initial_fraction)/total_cfus)
    
    cell_size = 1.7 #CHECK - could be different for different genotypes
    radii = np.sqrt(areas/math.pi)
    est_prob = sectors*cell_size/(2*math.pi*radii*initial_fraction)
    est_prob_err = est_prob*initial_fraction_err/initial_fraction

    genotype_array = np.tile(np.transpose(genotypes), 8)[0]

    pest = pd.DataFrame(columns = ['genotype', 'cm', 'p0_expected', 'p0_actual', 'p0_actual_err', 'replicate', 'pest', 'pest_err'])
    pest['genotype'] = genotype_array
    pest['cm'] = cm
    pest['p0_expected'] = p0
    pest['p0_actual'] = initial_fraction
    pest['p0_actual_err'] = initial_fraction_err
    pest['replicate'] = np.repeat(np.array(list(range(8))),12)
    pest['pest'] = est_prob
    pest['pest_err'] = est_prob_err
    pest['num_sectors'] = sectors
    pest_all = pest_all.append(pest)
```

```python
# Fast test
# Fitting linear relationship between s = -0.1 and 0.5, using all p0

filename = '/Users/qinqinyu/Documents/hallatscheck_lab/scripts/keio_collection/validation.csv'
df_validation = pd.read_csv(filename)
fits = pd.DataFrame(columns = ['genotype', 's0.05', 's0.1', 's0.2', 's0.05_err', 's0.1_err', 's0.2_err'])

s = [0.05, 0.1, 0.15, 0.2]

fig, ax = plt.subplots(3, 3, figsize = (10, 8))

i = 0
j = 0

for genotype in genotypes['genotype'].values:
    fitness = fitness_all[fitness_all['genotype']==genotype]
    fitness_tiled = np.tile(fitness['s'].values, 3)
    pest = pest_all[pest_all['genotype']==genotype]
    pest_p0_002 = pest[pest['p0_expected'] == 0.002]['pest'].values
    pest_p0_005 = pest[pest['p0_expected'] == 0.005]['pest'].values
    pest_p0_020 = pest[pest['p0_expected'] == 0.020]['pest'].values
    pest_ordered = np.concatenate((pest_p0_002, pest_p0_005, pest_p0_020))
    
    idx = np.logical_and(np.logical_and(fitness_tiled<0.5, fitness_tiled>-0.1), ~np.isnan(pest_ordered))

    x = fitness_tiled[idx]
    y = pest_ordered[idx]
    

#     plt.plot(x, y, label = genotype + ' data', marker = 'o', linestyle = '')

    if len(x)>0:
        z, cov = np.polyfit(x, y, 1, full = False, cov = True)
        zerr = np.sqrt(np.diag(cov))
        p = np.poly1d(z)
        ppp = np.poly1d(z + zerr)
        pmm = np.poly1d(z - zerr)
        ppm = np.poly1d(z + np.array([1, -1])*zerr)
        pmp = np.poly1d(z + np.array([-1, 1])*zerr)
        
        dpp = ppp(s)
        dmm = pmm(s)
        dpm = ppm(s)
        dmp = pmp(s)

        d = np.array([dpp, dmm, dpm, dmp])
        dmax = np.max(d, axis = 0)
        dmin = np.min(d, axis = 0)
        
        err = (dmax-dmin)/2
        
        xfit = np.linspace(-0.1, 0.5, 10)
        fit = pd.DataFrame([np.append(np.append(genotype, p(s)),err)], columns = ['genotype', 's0.05', 's0.1', 's0.15', 's0.2', 's0.05_err', 's0.1_err', 's0.15_err', 's0.2_err'])
        fits = fits.append(fit)
        
        if ~np.isnan(z[0]):
            ax[i,j].scatter(fitness['s'], pest_p0_002, label = '$p_0$ = 0.002')# + str(round(pest[pest['p0_expected'] == 0.002]['p0_actual'].values[0], 4)))
            ax[i,j].scatter(fitness['s'], pest_p0_005, label = '$p_0$ = 0.005')# + str(round(pest[pest['p0_expected'] == 0.005]['p0_actual'].values[0], 4)))
            ax[i,j].scatter(fitness['s'], pest_p0_020, label = '$p_0$ = 0.02')# + str(round(pest[pest['p0_expected'] == 0.020]['p0_actual'].values[0], 4)))

            ax[i,j].plot(xfit, p(xfit), 'k-', label = 'linear fit')

            ax[i,j].fill_between([-0.5, -0.1], 0, 1, color = 'gray', alpha = 0.2)
            ax[i,j].fill_between([0.5, 1.5], 0, 1, color = 'gray', alpha = 0.2)
            ax[i,j].set_xlim([-0.5, 1.5])
            ax[i,j].set_ylim([0, 1])
            ax[i,j].set_title(genotype)
    
            if (j+1)%3 == 0:
                i += 1
                j = 0
            else:
                j += 1
fits.reset_index(inplace = True, drop = True)

ax[2,1].set_xlabel('fitness')
ax[1,0].set_ylabel('establishment probability')

ax[1,2].legend(loc = (1.05, 0.15))
plt.tight_layout()

# plt.savefig('/Users/qinqinyu/Dropbox/keio_data/nonneutral_data/figures/all_genotypes_fit.png')
# plt.savefig('/Users/qinqinyu/Dropbox/keio_data/nonneutral_data/figures/all_genotypes_fit.pdf')
plt.show()
```

```python
def pest_theory(phi0, D, s):
    return erf(phi0/np.sqrt(4*D)) + (1-np.log((s*phi0)/(2*D))*phi0/np.sqrt(4*math.pi*D))*(phi0*s)/(2*D)
```

```python
# Including error bars from p0
# Fitting linear relationship between s = -0.1 and 0.5, using all p0

filename = '/Users/qinqinyu/Documents/hallatscheck_lab/scripts/keio_collection/validation.csv'
df_validation = pd.read_csv(filename)
fits = pd.DataFrame(columns = ['genotype', 's0.05', 's0.1', 's0.2', 's0.05_err', 's0.1_err', 's0.2_err'])

s = [0.05, 0.1, 0.15]

for genotype in genotypes['genotype'].values:
    fitness = fitness_all[fitness_all['genotype']==genotype]
    fitness_tiled = np.tile(fitness['s'].values, 3)
    pest = pest_all[pest_all['genotype']==genotype]
    pest_p0_002 = pest[pest['p0_expected'] == 0.002]['pest'].values
    pest_p0_005 = pest[pest['p0_expected'] == 0.005]['pest'].values
    pest_p0_020 = pest[pest['p0_expected'] == 0.020]['pest'].values
    pest_ordered = np.concatenate((pest_p0_002, pest_p0_005, pest_p0_020))
    
    pest_err_p0_002 = pest[pest['p0_expected'] == 0.002]['pest_err'].values
    pest_err_p0_005 = pest[pest['p0_expected'] == 0.005]['pest_err'].values
    pest_err_p0_020 = pest[pest['p0_expected'] == 0.020]['pest_err'].values
    pest_err_ordered = np.concatenate((pest_err_p0_002, pest_err_p0_005, pest_err_p0_020))
    
    idx = np.logical_and(np.logical_and(np.logical_and(fitness_tiled<0.5, fitness_tiled>-0.1), ~np.isnan(pest_ordered)), ~np.isnan(pest_err_ordered))

    x = fitness_tiled[idx]
    y = pest_ordered[idx]
    yerr = pest_err_ordered[idx]
    yerr = yerr + 10**-4
#     print(yerr)
#     plt.figure()
#     plt.errorbar(x, y, yerr = yerr, label = genotype + ' data', marker = 'o', linestyle = '')

    if len(x)>0:
        z, cov = np.polyfit(x, y, 1, w = 1/yerr, full = False, cov = True)
        zerr = np.sqrt(np.diag(cov))
        p = np.poly1d(z)
        ppp = np.poly1d(z + zerr)
        pmm = np.poly1d(z - zerr)
        ppm = np.poly1d(z + np.array([1, -1])*zerr)
        pmp = np.poly1d(z + np.array([-1, 1])*zerr)
        
        dpp = ppp(s)
        dmm = pmm(s)
        dpm = ppm(s)
        dmp = pmp(s)

        d = np.array([dpp, dmm, dpm, dmp])
        dmax = np.max(d, axis = 0)
        dmin = np.min(d, axis = 0)
        
        err = (dmax-dmin)/2
#         print()
        plt.plot(x, p(x), label = genotype + ' fit')
        fit = pd.DataFrame([np.append(np.append(genotype, p(s)),err)], columns = ['genotype', 's0.05', 's0.1', 's0.2', 's0.05_err', 's0.1_err', 's0.2_err'])
        fits = fits.append(fit)
fits.reset_index(inplace = True, drop = True)

plt.ylim([0, 10**-4])
plt.xlabel('fitness')
plt.ylabel('establishment probability')

plt.legend()
plt.show()

merged = df_validation.merge(fits, on = 'genotype')
plt.errorbar(merged['MSD (L = 50um)'].values.astype('float'), merged['s0.05'].values.astype('float'), yerr = merged['s0.05_err'].values.astype('float'), xerr = merged['MSD ste (L = 50um)'].values.astype('float'), linestyle = '', marker = 'o', label = 's = 0.05')
plt.errorbar(merged['MSD (L = 50um)'].values.astype('float'), merged['s0.1'].values.astype('float'), yerr = merged['s0.1_err'].values.astype('float'), xerr = merged['MSD ste (L = 50um)'].values.astype('float'), linestyle = '', marker = 'o', label = 's = 0.1')
plt.errorbar(merged['MSD (L = 50um)'].values.astype('float'), merged['s0.2'].values.astype('float'), yerr = merged['s0.2_err'].values.astype('float'), xerr = merged['MSD ste (L = 50um)'].values.astype('float'), linestyle = '', marker = 'o', label = 's = 0.2')

plt.legend()
# plt.xlim([0, 3])
# plt.ylim([10**-3, 0.2])
# plt.yscale('log')
plt.xlabel('mean squared displacement, MSD \n (L = 50$\mu m$) [$\mu m^2$]')
plt.ylabel('establishment probability, $p_{est}$')
plt.show()
```
